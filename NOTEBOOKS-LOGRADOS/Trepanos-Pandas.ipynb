##### para devolver la media del costo de cada tipo de trepano

```python
df_group = df.groupby('TIPO')['COSTO'].mean()
```

```python
df_group.name='media_del_costo'
```

```python
df_group
```

##### para crear otra columna ahora con la media del costo

```python
df_join = df.join(df_group, on=['TIPO'], how='inner')
```
