```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def h(x):
    return np.cos(x)
```

```python
x=np.linspace(0,10,100)
```

```python
plt.plot(x,h(x))
plt.xlabel('tiempo - espacio')
#plt.xlabel('posicion')
plt.title('Funcion Coseno - Daiana Daher')
plt.legend(loc=1)
plt.grid()
```
