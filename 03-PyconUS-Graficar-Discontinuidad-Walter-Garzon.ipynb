```python
import matplotlib.pyplot as Funcion
import numpy as np
```

```python
X = np.linspace(-0.1, 0.1, 35)
```

```python
X
```

```python
print(X)
```

```python
Y = np.sin(X)/X
```

## Función discontinua con numpy

La función *sin(X)/X* es discontinua en x=0, no está definida, y **Python numpy** informa el error, pero lo grafica igualmente 

```python
Y
```

```python
# Create a figure of size 8x6 inches, 80 dots per inch
Funcion.figure(figsize=(14, 10), dpi=80)
Funcion.grid(True)
Funcion.plot(X,Y,"bo")
```
